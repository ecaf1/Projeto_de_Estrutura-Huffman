```python
import pandas as pd
import matplotlib.pyplot as plt

# Lê o arquivo CSV
df = pd.read_csv("dados.csv")
df1= pd.read_csv("dados1.csv")

# Seleciona as colunas desejadas
x = df["tam"]
y1 = df["elastic_FP"]
y2 = df['elastic_heap']

y3 = df1['elastic_FP']
y4 = df1['elastic_heap']

# Plota o gráfico
fig, axs = plt.subplots(2, 2, figsize=(15,6))
fig.suptitle("ENQUEUE CRESCENTE", fontsize=22)

axs[0, 0].scatter(x, y1)
axs[0, 0].set_xlabel("Tamanho")
axs[0, 0].set_ylabel("Interações")
axs[0, 0].set_title("Enqueue Fila de Prioridade")

axs[0, 1].scatter(x, y2)
axs[0, 1].set_xlabel("Tamanho")
axs[0, 1].set_ylabel("Interações")
axs[0, 1].set_title("Enqueue HEAP")

axs[1, 0].scatter(x, y3)
axs[1, 0].set_xlabel("Tamanho")
axs[1, 0].set_ylabel("Interações")
axs[1, 0].set_title("Enqueue Fila de Prioridade")

axs[1, 1].scatter(x, y4)
axs[1, 1].set_xlabel("Tamanho")
axs[1, 1].set_ylabel("Interações")
axs[1, 1].set_title("Enqueue HEAP")

plt.subplots_adjust(top=0.85, bottom=0.1, left=0.1, right=0.9, hspace=0.5, wspace=0.3)
plt.show()
```
